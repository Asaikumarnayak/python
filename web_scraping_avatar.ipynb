```python
import pandas as pd
import numpy as np
import requests as r
from bs4 import BeautifulSoup as b
```

```python
try:
  re=r.get('https://www6.f2movies.to/home')
except exception as e:
  print("error")  
```

```python
s=b(re.text,'html.parser')
```

```python
s
```

```python
a=s.find('section',{'class':'block_area'}).findAll('div',{'class':'film-detail'})
```

```python
dic=defaultdict(list)
```

```python
for i in a:
  title=i.h3.text
  release=i.find('span',{'class':'fdi-item'}).text
  # duration=i.find('span',{'class':'fdi-item fdi-duration'}).text
  type=i.find('span',{'class':'float-right fdi-type'}).text

  dic['title'].append(title)
  dic['release'].append(release)
  dic['duration'].append(duration)
  dic['type'].append(type)
```

```python
dic
```

```python
df=pd.DataFrame(dic)
```

```python
df
```

```python
df.to_csv('movie.csv',index=False)
```
